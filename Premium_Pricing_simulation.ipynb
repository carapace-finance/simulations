```python
#setting up constant values
time_to_expiration_constant = 1
protection_amount_constant = 100
curvature_constant = 1/20
minimum_risk_factor_constant = 0.03
leverage_ratio_min = 0.05
leverage_ratio_max = 0.25
min_premium = 0.02
leverage_ratio_constant = 0.15
risk_premium_constant = 0.1
borrower_apy_constant = 17
```

```python
import math
#premium = base_premium + risk_premium
def base_premium( 
            protection_amount  = protection_amount_constant, 
            time_to_expiration = time_to_expiration_constant, 
            curvature          = curvature_constant,
            leverage_ratio     = leverage_ratio_constant,
           ):
    
    risk_factor      = ((leverage_ratio_max - leverage_ratio)/(leverage_ratio-leverage_ratio_min)) * curvature
    base_premium     = max((1-pow(math.e,-time_to_expiration * risk_factor)), min_premium ) * protection_amount
    return base_premium
```

```python
print(base_premium(leverage_ratio = 0.15))
```

```python
# importing the required modules
from matplotlib import pyplot as plt
import numpy as np
import math
curvature_constants = [1/20, 1/25, 1/30, 1/40, 1/ 50, 1/60, 1/70, 1/80, 1/90, 1/100 ]
# setting the x - coordinates
for curvature in curvature_constants:
    leverage_ratios = np.arange(0.1, 0.2 , 0.001)
    base_premiums = [base_premium(leverage_ratio = leverage_ratio, curvature = curvature) for leverage_ratio in leverage_ratios ]
    print(curvature_constant)
    # plotting the points
    plt.plot(leverage_ratios, base_premiums)
    plt.xlabel("leverage ratio")
    plt.ylabel("Base Premium payment")
    plt.show()   
```

```python

def risk_premium(
        risk_premium_percent = risk_premium_constant,
        borrower_apy         = borrower_apy_constant,
    ):
    risk_premium = risk_premium_percent * borrower_apy
    return risk_premium

```

```python
def premium( 
        protection_amount    = protection_amount_constant, 
        time_to_expiration   = time_to_expiration_constant, 
        curvature            = curvature_constant,
        leverage_ratio       = leverage_ratio_constant,
        risk_premium_percent = risk_premium_constant,
        borrower_apy         = borrower_apy_constant,
       ):
    
    Base_premium =  base_premium(
                        protection_amount  = protection_amount, 
                        time_to_expiration = time_to_expiration, 
                        curvature          = curvature,
                        leverage_ratio     = leverage_ratio,
                    )
    
    Risk_premium =  risk_premium(
                        risk_premium_percent = risk_premium_percent,
                        borrower_apy         = borrower_apy,
                    )
    
    premium = Base_premium + Risk_premium
    return premium
```
